```python
img_rgb = cv2.imread('/content/ssesesese.png')
# img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2GRAY)
template = cv2.imread('/content/trterter.png')
print(template.shape[::-1])
_,w, h = template.shape[::-1]
res = cv2.matchTemplate(img_rgb,template,cv2.TM_CCOEFF_NORMED)
threshold = 0.874
loc = np.where( res >= threshold)
count=0
tr=[]
for pt in zip(*loc[::-1]):
  cv2.rectangle(img_rgb, pt, (pt[0]+w, pt[1]+h), (0,255,0), 2)
  count+=1
  tr.append([pt[0],pt[1]])
cv2_imshow(img_rgb)
print(count)
```

```python
img_rgb = cv2.imread('/content/ssesesese.png')
# img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2GRAY)
template = cv2.imread('/content/rec5.png')
print(template.shape[::-1])
_,w, h = template.shape[::-1]
res = cv2.matchTemplate(img_rgb,template,cv2.TM_CCOEFF_NORMED)
threshold = 0.82
loc = np.where( res >= threshold)
count=0
recs=[]
for pt in zip(*loc[::-1]):
  cv2.rectangle(img_rgb, pt, (pt[0]+w, pt[1]+h), (0,0,0), 2)
  count+=1
  recs.append([pt[0],pt[1]])
cv2_imshow(img_rgb)
print(count)
```

```python
import matplotlib.pyplot as plt
img_rgb = cv2.imread('/content/ssesesese.png')
Z = np.vstack((tr,recs))
# convert to np.float32
Z = np.float32(Z)
# define criteria and apply kmeans()
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
ret,label,center=cv2.kmeans(Z,2,None,criteria,10,cv2.KMEANS_RANDOM_CENTERS)
# Now separate the data, Note the flatten()
A = Z[label.ravel()==0]
B = Z[label.ravel()==1]
# Plot the data
# plt.subplot(1,2,1)
plt.gca().invert_yaxis()
plt.imshow(img_rgb,'brg')
plt.scatter(A[:,0],A[:,1])
plt.scatter(B[:,0],B[:,1],c = 'r')
plt.scatter(center[:,0],center[:,1],s = 80,c = 'y', marker = 's')
plt.xlabel('Height'),plt.ylabel('Weight')
plt.show()
```

```python
resp=[]

for i in range(len(recs_ans)):
  resp.append(float(0))
for i in range(len(tr_ans)):
  resp.append(float(1))
print(resp)
knn = cv2.ml.KNearest_create()
print(len(resp),len(trainData))
resp=np.array(resp)
resp=resp.astype(np.float32)
print(type(resp[0]))
print(type(trainData[0][1]))
knn.train(trainData, cv2.ml.ROW_SAMPLE, resp)
ring=ring.astype(np.float32)
ret, results, neighbours, dist = knn.findNearest(ring, 9)
print("result: ", results); print("neighbours: ", neighbours)
print("distance: ", dist)
```

```python
ring_tr=[]
ring_rec=[]
for i in range(len(results)):
  if(int(results[i])==0):
    ring_tr.append(ring[i])
  else:
    ring_rec.append(ring[i])
ring_tr=np.array(ring_tr)
ring_rec=np.array(ring_rec)
```
